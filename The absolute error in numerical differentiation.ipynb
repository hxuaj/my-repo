## This is the plot that shows the absolute error in numerical differentiation

https://en.wikipedia.org/wiki/Numerical_differentiation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x_range = np.array([10, 0.1, 0.001])
h = np.array([np.power(0.1, i*0.1) for i in range(16*10, -1, -1)])

for x in x_range:
    #h = np.array([np.power(0.1, i) for i in range(16, 0, -1)])
    y = np.abs((np.power((h+x),3) - np.power(x,3))/h - 3*np.power(x, 2))
    plt.plot(np.log10(h), np.log10(y))
    
x_centered = 0.1
y_centered = np.abs((np.power((x_centered+h),3)
                     - np.power((x_centered-h),3))/(2*h)
                    - 3*np.power(x_centered, 2))
plt.plot(np.log10(h), np.log10(y_centered))
    
plt.xticks(np.linspace(-16,0,5), [r'1e{}'.format(i) for i in range(-16, 1, 4)])
plt.yticks(np.linspace(-16,3,20), [r'1e{}'.format(i) for i in range(-16, 4)])
plt.legend(['x = {}'.format(i) for i in x_range] + ['centered difference, x = 0.1'])
plt.title('The absolute error in numerical differentiation')
plt.xlabel('h')
plt.ylabel('Error')
plt.show()
```
