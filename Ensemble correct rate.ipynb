## This is the plot that shows the ensemble correct rate

```python
import matplotlib.pyplot as plt
import numpy as np

def find_error(e, n):
    #e:error rate for each classifier, n: the number of classifier#
    
    total_error_rate = 0
    for i in range((n//2)+1, n+1):
        total_error_rate += np.power(e, i) * np.power(1-e,n-i)
    return total_error_rate
x = np.arange(0,1,0.01)
y = find_error(x, n = 5)

plt.title("Ensemble Correct Rate")
plt.plot(1-x, 1-y)
plt.xlabel("Single model accuracy")
plt.ylabel("Total accuracy")
plt.show()
```
